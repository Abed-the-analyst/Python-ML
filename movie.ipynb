```python
# Import libraries

import pandas as pd
import numpy as np 
import seaborn as sns
from sklearn.impute import SimpleImputer
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts the configuration of the plots we will create


# read the data
df = pd.read_csv(r'C:\Users\abedw\Jupyterr\movies\movies.csv')
```

```python
#let's look at the data

df.head()
```

```python
#let's see if there's any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}  - {}%'.format(col, pct_missing))
```

```python
missing_val_count_by_column = (df.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```python
df_cleaned= df.fillna(method='bfill', axis=0)
missing_val_count_by_column = (df_cleaned.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```python
# data types for our columns

df.dtypes
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 20)
```

```python
# drop any duplicate

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# scatter plot with budget Vs gross


plt.scatter(x=df['budget'], y=df['gross'])

plt.title('budget vs gross earnings')

plt.xlabel('gross earnings')

plt.ylabel('budget for film')

plt.show()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df_cleaned, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# let's start looking for correlation 
```

```python
df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# high correlation in budget and gross

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric features')

plt.xlabel('movie features')

plt.ylabel('movie features')

plt.show()
```

```python
# look at company 

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric features')

plt.xlabel('movie features')

plt.ylabel('movie features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
